```python
import pandas as pd
```

```python
df = pd.read_csv("weather_data.csv", parse_dates=["day"]) #Change str to date_form
df
```

```python
type(df.day[0])
```

```python
df.columns
```

```python
df.set_index('day', inplace=True)
df
```

```python
df.fillna(0)
df
```

```python
df.fillna({
    'temperature': df.temperature.mean(),
    'windspeed': df.windspeed.mean(),
    'event': 'No Event'
})
```

```python
df.fillna(method='ffill') #Passing the value from the previous day to the next day
```

```python
df.fillna(method='bfill',axis='columns') #backward fill
```

```python
df.interpolate()
```

```python
df.dropna(how="all")
```

```python
df.dropna(thresh=2)
```
